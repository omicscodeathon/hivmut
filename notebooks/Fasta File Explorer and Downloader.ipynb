# Initial Download of the Genomes

```python
from Bio import Entrez, SeqIO
import os
import threading
from tqdm import tqdm

accessions = df["Accession"].dropna().unique().tolist()

len(accessions)
```

```python
output_dirs = "../output/hiv_fasta_files"
os.makedirs(output_dirs, exist_ok=True)
```

```python
def download_fasta(accession, output_dir):
    Entrez.email = "halleluyaholudele@gmail.com"
    output_path = os.path.join(output_dir, f"{accession}.fasta")

    if os.path.exists(output_path):
        return

    try:
        handle = Entrez.efetch(
            db="nucleotide", id=accession, rettype="fasta", retmode="text"
        )
        record = SeqIO.read(handle, "fasta")
        handle.close()

        SeqIO.write(record, output_path, "fasta")
    except Exception as e:
        print(f"Failed to download file with accession: {accession} \n Error: {e}")
```

```python
def download_worker(accessions, output_dir):
    for accession in accessions:
        download_fasta(accession, output_dir)


def parallel_download(accessions, output_dir, num_threads=4):
    chunks = [accessions[i::num_threads] for i in range(num_threads)]

    threads = []
    for chunk in chunks:
        thread = threading.Thread(target=download_worker, args=(chunk, output_dir))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()


num_threads = 4

# Downloading with progress bar
with tqdm(total=len(accessions), desc="Downloading FASTA files", unit="file") as pbar:
    for accession in accessions:
        download_fasta(accession, output_dirs)
        pbar.update(1)

parallel_download(accessions, output_dirs, num_threads)
```
