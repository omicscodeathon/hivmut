```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from rich import print

pd.options.display.max_columns = None
```

```python
data1_path = '../output/Nigerian-Drug-Resistance-Analysis/analysis-reports/resistanceSummaries.csv'
data2_path = '../output/Nigerian-Drug-Resistance-Analysis/analysis-reports 2/resistanceSummaries.csv'
data3_path = '../output/Nigerian-Drug-Resistance-Analysis/analysis-reports 3/resistanceSummaries.csv'
data4_path = '../output/Nigerian-Drug-Resistance-Analysis/analysis-reports 4/resistanceSummaries.csv'
```

```python
df1 = pd.read_csv(data1_path)
df2 = pd.read_csv(data2_path)
df3 = pd.read_csv(data3_path)
df4 = pd.read_csv(data4_path)
```

```python
df = pd.concat([df1, df2, df3, df4], ignore_index=True)
```

```python
df.describe()
```

```python
df.head()
```

```python
df['Accession ID'] = df['Sequence Name'].str.split(' ').str[0]
```

```python
df.columns
```

```python
df['Genes'].value_counts()
```

```python
# Histograms for resistance scores
plt.figure(figsize=(12, 8))
sns.histplot(df['ATV/r Score'], kde=True)
plt.title('Distribution of ATV/r Resistance Scores')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df[['ATV/r Score', 'DRV/r Score', 'FPV/r Score', 'LPV/r Score']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Multiple Drug Scores')
plt.show()
```

```python
df['Algorithm Date Modified'] = pd.to_datetime(df['Algorithm Date'])
plt.figure(figsize=(12, 8))
sns.lineplot(x='Algorithm Date Modified', y='ATV/r Score', data=df)
plt.title('Trend of ATV/r Resistance Scores Over Time')
plt.show()
```
