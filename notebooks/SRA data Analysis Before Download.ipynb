```python
illumina_filtered_df = df[df['Platform'] == 'ILLUMINA']
illumina_filtered_df['Platform'].value_counts()
```

```python
nano_filtered_df = df[df['Platform'] == 'OXFORD_NANOPORE']
```

```python
df['geo_loc_name_country'].value_counts()
```

```python
runs = illumina_filtered_df['Run']
nano_runs = nano_filtered_df['Run']
```

```python
with open('../analysis/illumina_accessions.txt', 'w') as f:
    for run in runs:
        f.write(f'{run}\n')
print('Accessions written to the illumina_accessions.txt file')
```

```python
with open('../analysis/nanopore_accessions.txt', 'w') as f:
    for run in nano_runs:
        f.write(f'{run}\n')
print('Accessions written to the nanopore_accessions.txt file')
```
